```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

# qaysi turdagi vrdan qancha vaqtdan beri foydalashi 

```python
sns.stripplot(data=df,
              x='Duration',
              y='VRHeadset',
              jitter=True,
              hue='VRHeadset')
```

# qaysi turdagi vrdan qancha yoshdagilar foydalashi 

```python
sns.swarmplot(data=df,
              x='Age',
              y='VRHeadset',
              hue='VRHeadset')
```

```python
sns.boxplot(data=df,
              x='Age',
              y='VRHeadset',
              hue='VRHeadset')
```

```python
sns.boxenplot(data=df,
              x='Age',
              y='VRHeadset',
              hue='VRHeadset')
```

```python
sns.violinplot(data=df,
              x='Age',
              y='VRHeadset',
              hue='VRHeadset')
```

```python
sns.regplot(data=df,
            y='Duration',
            x='Age',
            order=2)
```

# yoshi va qaysi turdagi vrda ortacha qancha vaqt ishlatgani

```python
df_crosstab = pd.crosstab(df['Age'], df['VRHeadset'], values=df['Duration'], aggfunc="mean").round(2)
df_crosstab.iloc[:10]
```

```python
sns.heatmap(df_crosstab.iloc[:10], annot=True)
```

```python
sns.heatmap(df_crosstab.iloc[:10].astype(int), annot=True, fmt="d",
            cmap="YlGnBu", cbar=False, linewidths=.5)
```
