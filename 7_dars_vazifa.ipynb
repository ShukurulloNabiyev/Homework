```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
medal = pd.read_csv('medals_by_country_2016.csv')
df = pd.read_csv('wind_dataset.csv')
df.head()
```

# temperaturani korsatgichlari

```python
fig, ax = plt.subplots()
ax.plot(df['DATE'].iloc[:12], df['T.MAX'].iloc[:12])
ax.set_ylabel('Temperature')
ax.set_xlabel('Day')
ax.set_xticklabels(df['DATE'], rotation=45)
fig.set_size_inches(10,5)
```

# shamol va yomgirni korsatgichlar

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(df['DATE'].iloc[:12], df['RAIN'].iloc[:12])
ax[0].set_ylabel('RAIN')
ax[0].set_xlabel('Day')
ax[0].set_xticklabels(df['DATE'], rotation=45)
fig.set_size_inches(12,6)

ax[1].plot(df['DATE'].iloc[:12], df['WIND'].iloc[:12])
ax[1].set_ylabel('WIND')
ax[1].set_xlabel('Day')
ax[1].set_xticklabels(df['DATE'], rotation=45)
```

# harorat va yomgirni yogishini solishtirish

```python
plt.style.use("ggplot") 
fig, ax = plt.subplots()
ax.scatter( df['T.MAX'], df['RAIN'])
ax.set_ylabel('Temperature(C)')
ax.set_xlabel('Rain dagree')
```

# Jadvalda korinib turibdiku havo issiq bolganda yomgir kam yogadi

```python
fig, ax = plt.subplots()
ax.scatter(df['WIND'], df['RAIN'])
```

# temperatura shamolning ortachasi va stdisi

```python
fig, ax = plt.subplots()
ax.bar('Temperature', df['T.MAX'].mean(), yerr=df['T.MAX'].std())
ax.bar('Wind', df['WIND'].mean(), yerr=df['WIND'].std())
ax.bar('Rain', df['RAIN'].mean(), yerr=df['RAIN'].std())
```

```python
fig, ax = plt.subplots()
ax.bar(medal['Country'], medal['Gold'], label='Gold')
ax.bar(medal['Country'], medal['Silver'], bottom= medal['Gold'], label='Silver')
ax.bar(medal['Country'], medal['Bronze'], bottom= medal['Gold'] + medal['Silver'], label='Bronze')
ax.set_xticklabels(medal['Country'], rotation=45)
ax.legend()
plt.show()
```
