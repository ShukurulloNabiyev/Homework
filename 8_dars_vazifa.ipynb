```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df.head()
```

![Alt text](image.png)

# pingivinlarni qanotini uzunligi bn tumshigini uzunligini solishtiramiz

```python
sns.scatterplot(x="flipper_length_mm",
                y="bill_length_mm",
                data=df)
```

# korinib turibdiki qanoti uzunlarni tumshigi ham uzun bolarkan

# endi pingvin massa bn qanotining uzunligini solishtiramiz

```python
sns.scatterplot(x="flipper_length_mm",
                y="body_mass_g",
                data=df)
```

```python
sns.scatterplot(x="flipper_length_mm",
                y="body_mass_g",
                data=df,
                hue='sex')
```

# orollar miqiyosida pingvinlarni ajratib chiqamiz

```python
sns.scatterplot(x="flipper_length_mm",
                y="body_mass_g",
                data=df,
                hue='island',
                size='island')
```

# jadvalda korganizdek eng ogir va qanoti uzun pingvinlar torgerson orolida ekan 

# pingvinlarni turlariga bolib chiqamiz

```python
sns.scatterplot(x="flipper_length_mm",
                y="body_mass_g",
                data=df,
                hue='species',
                size='species')
```

# pingvinlarni soni

```python
sns.countplot(x='species',
              data=df,
              hue='species')
```

```python
sns.countplot(x='species',
              data=df,
              hue='sex')
```

```python
sns.catplot(x='species',
            y='flipper_length_mm',
            data=df,
            kind='bar',
            hue='species')
```

# pingvinlarni ogirligi boyicha

```python
sns.catplot(x='species',
            y='body_mass_g',
            data=df,
            kind='box',
            hue='species')
```
