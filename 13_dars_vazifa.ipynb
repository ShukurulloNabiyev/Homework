```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
new_df = df[['displacement', 'horsepower', 'weight']]

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```

# Standardisation

```python
new1_df = (new_df - new_df.mean()) / new_df.std()

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new1_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new1_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new1_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```

# MinMaxScaling

```python
new1_df = (new_df - new_df.min()) / (new_df.max() - new_df.min())

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new1_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new1_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new1_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```

# Mean Normalisation

```python
new1_df = (new_df - new_df.mean()) / (new_df.max() - new_df.min())

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new1_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new1_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new1_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```

# MaxAbsScaling

```python
new1_df = (new_df) / new_df.abs().max()

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new1_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new1_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new1_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```

# Robust Scaling

```python
new1_df = (new_df - new_df.median()) / (new_df.quantile(0.75) - new_df.quantile(0.25))

fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)
sns.histplot(data=new1_df['displacement'], kde=True, ax=ax0)
sns.histplot(data=new1_df['horsepower'], kde=True, ax=ax1)
sns.histplot(data=new1_df['weight'], kde=True, ax=ax2)

sns.set_style('whitegrid')

fig.set_size_inches(12,6)
```
