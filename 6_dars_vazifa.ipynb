```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
olympics = pd.read_csv('athletes.csv')
olympics.head()
olympics.drop_duplicates
nima = pd.DataFrame()
```

# har bir sport turida qanchadan odam qatnashayotganini topish

```python
sport = olympics.groupby('sport')['id'].count().to_dict()
sports_name = list(sport.keys())[:10]
humans_count = list(sport.values())[:10]
```

```python
fig, ax = plt.subplots()
ax.bar(sports_name, humans_count)
ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Humans count')
```

# sportchilarni jinsiga qarab ajratib chiqish

```python
olympics.groupby(['sport', 'sex'])['sex'].count()
```

```python
gender = olympics.groupby(['sport', 'sex'])['sex'].count().tolist()[:20]
female = [gender[i] for i in range(len(gender)) if i % 2 == 0]
male = [gender[i] for i in range(len(gender)) if i % 2 != 0]

fig, ax = plt.subplots()
ax.bar(sports_name, male, label='Erkaklar')
ax.bar(sports_name, female, bottom=male, label= 'Ayollar')

ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Humans count')
ax.legend()
```

# har bir sport turida qancha oltin medal olingani topish

```python
gold_medals = olympics.groupby('sport')['gold'].sum().to_dict()
gold_medals = list(gold_medals.values())[:10]

fig, ax = plt.subplots()
ax.bar(sports_name, gold_medals)
ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Medal count')
```

# erkaklar va ayollar nechtadan oltin medal olishgani

```python
olympics.groupby(['sport', 'sex'])['gold'].sum()
```

```python
gold = olympics.groupby(['sport', 'sex'])['gold'].sum().tolist()[:20]

female_gold = [gold[i] for i in range(len(gold)) if i % 2 == 0]
male_gold = [gold[i] for i in range(len(gold)) if i % 2 != 0]

fig, ax = plt.subplots()
ax.bar(sports_name, male_gold, label='Erkaklar')
ax.bar(sports_name, female_gold, bottom=male_gold, label= 'Ayollar')

ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Medals count')
ax.legend()
```

# nechtadan medallar qolga kiritilganligi

```python
gold_medals = olympics.groupby('sport')['gold'].sum().to_dict()
gold_medals = list(gold_medals.values())[:10]

silver_medals = olympics.groupby('sport')['silver'].sum().to_dict()
silver_medals = list(silver_medals.values())[:10]

bronze_medals = olympics.groupby('sport')['bronze'].sum().to_dict()
bronze_medals = list(bronze_medals.values())[:10]


fig, ax = plt.subplots()

ax.bar(sports_name, gold_medals, label= 'Gold')
ax.bar(sports_name, silver_medals, bottom= gold_medals, label='Silver')
ax.bar(sports_name, bronze_medals, bottom= np.array(gold_medals) + np.array(silver_medals), label='Bronze')

ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Medal count')
ax.legend()
```

# erkaklar va ayollarni melarini masklari

```python
silver = olympics.groupby(['sport', 'sex'])['silver'].sum().tolist()[:20]

female_silver = [silver[i] for i in range(20) if i % 2 == 0]
male_silver = [silver[i] for i in range(20) if i % 2 != 0]

bronze = olympics.groupby(['sport', 'sex'])['silver'].sum().tolist()[:20]

female_bronze = [bronze[i] for i in range(20) if i % 2 == 0]
male_bronze = [bronze[i] for i in range(20) if i % 2 != 0]
```

# erkaklar umumiy qancha medal olgani

```python
fig, ax = plt.subplots()

ax.bar(sports_name, male_gold, label= 'Gold')
ax.bar(sports_name, male_silver, bottom= male_silver, label='Silver')
ax.bar(sports_name, male_bronze, bottom= np.array(male_gold) + np.array(male_silver), label='Bronze')

ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Medal count')
ax.legend()
```

# ayollar umumiy qancha medal olgani

```python
fig, ax = plt.subplots()

ax.bar(sports_name, female_gold, label= 'Gold')
ax.bar(sports_name, female_silver, bottom= female_silver, label='Silver')
ax.bar(sports_name, female_bronze, bottom= np.array(female_gold) + np.array(female_silver), label='Bronze')

ax.set_xticklabels(sports_name, rotation=60)
ax.set_xlabel('Sports')
ax.set_ylabel('Medal count')
ax.legend()
```

# ayol suzuvchilar va basketbolchilarning vazni 

```python
womens_aquatics = olympics[(olympics['sex'] == 'female') & (olympics['sport'] == 'aquatics')]
womens_basketball = olympics[(olympics['sex'] == 'female') & (olympics['sport'] == 'basketball')]
```

```python
fig, ax = plt.subplots()  
ax.bar("Aquatics", womens_aquatics["weight"].mean())  
ax.bar("Baskatball", womens_basketball["weight"].mean())  
ax.set_ylabel("Weight (kg)")
```

```python
fig, ax = plt.subplots()  
ax.bar("Aquatics", womens_aquatics["weight"].mean(), yerr=womens_aquatics["weight"].std() )  
ax.bar("Baskatball", womens_basketball["weight"].mean(), yerr=womens_basketball["weight"].std())  
ax.set_ylabel("Weight (kg)")
```

```python
fig, ax = plt.subplots()
ax.hist(womens_aquatics['height'] * 100, label= 'Aquatics')
ax.hist(womens_basketball['height'] * 100, label= 'Basketball')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.hist(womens_aquatics['height'] * 100, label= 'Aquatics', histtype='step')
ax.hist(womens_basketball['height'] * 100, label= 'Basketball', histtype='step')
ax.legend()
```

```python
fig, ax = plt.subplots()  
ax.boxplot([womens_aquatics["height"] * 100, womens_basketball["height"] * 100])  
ax.set_xticklabels(["Aquatics", "Baskatball"])  
ax.set_ylabel("Height (cm)")
plt.show()
```
