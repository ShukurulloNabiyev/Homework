```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv('https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv')
df.head().T
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.select_dtypes('object')
```

# objectdagi numerik ustunlarni oz holiga keltirish

```python
df['COMPL_RPY_5YR_RT'] = pd.to_numeric(df['COMPL_RPY_5YR_RT'], errors='coerce')
df['DEBT_MDN'] = pd.to_numeric(df['DEBT_MDN'], errors='coerce')
df['MEDIAN_HH_INC'] = pd.to_numeric(df['MEDIAN_HH_INC'], errors='coerce')
```

```python
df.info()
```

# nullarni o'rniga mean bn medianni qoyib oldingi holi bn solishtirish 

```python
df['COMPL_RPY_5YR_RT_median'] = df.groupby(['REGION', 'Ownership'])['COMPL_RPY_5YR_RT'].transform(lambda x: x.fillna(x.median()))
df['COMPL_RPY_5YR_RT_mean'] = df.groupby(['REGION', 'Ownership'])['COMPL_RPY_5YR_RT'].transform(lambda x: x.fillna(x.mean()))

df['DEBT_MDN_median'] = df.groupby(['REGION', 'Ownership'])['DEBT_MDN'].transform(lambda x: x.fillna(x.median()))
df['DEBT_MDN_mean'] = df.groupby(['REGION', 'Ownership'])['DEBT_MDN'].transform(lambda x: x.fillna(x.mean()))
```

```python
fig, (ax0, ax1) = plt.subplots(1,2)
df['COMPL_RPY_5YR_RT'].plot(kind='kde', ax=ax0,  color='blue',  label='Numeric')
df['COMPL_RPY_5YR_RT_median'].plot(kind='kde', ax=ax0, color='red',  label='Median')
df['COMPL_RPY_5YR_RT_mean'].plot(kind='kde', ax=ax0, color='green',  label='Mean')


df['DEBT_MDN'].plot(kind='kde', ax=ax1,  color='blue',  label='Numeric')
df['DEBT_MDN_median'].plot(kind='kde', ax=ax1, color='red',  label='Median')
df['DEBT_MDN_mean'].plot(kind='kde', ax=ax1, color='green',  label='Mean')

ax0.legend()
ax1.legend()

ax0.set_title('COMPL_RPY_5YR_RT')
ax1.set_title('DEBT_MDN')


fig.set_size_inches(12,6)
fig.tight_layout()
plt.show()
```

# tutionni nullarni orniga mean medianni qoyib jadvalini koramiz

```python
df['Tuition_median'] = df.groupby(['Regions', 'Degree_Type'])['Tuition'].transform(lambda x: x.fillna(x.median()))
df['Tuition_mean'] = df.groupby(['Regions', 'Degree_Type'])['Tuition'].transform(lambda x: x.fillna(x.mean()))
```

```python
fig, ax0 = plt.subplots(figsize=(8,6))

df['Tuition'].plot(kind='kde', ax=ax0,  color='blue',  label='Numeric')
df['Tuition_median'].plot(kind='kde', ax=ax0, color='red',  label='Median')
df['Tuition_mean'].plot(kind='kde', ax=ax0, color='green',  label='Mean')

ax0.legend()
plt.show()
```
